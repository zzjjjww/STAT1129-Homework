```python
#basic functions + additional functions with hiding the first person's choice and giving the option to restart
import getpass
from enum import IntEnum

class Action(IntEnum):
    rock = 0
    paper = 1
    scissors = 2

def get_player1_name():
    global player1_name
    player1_name = input("Player 1, please enter your name: ")
    return player1_name

def get_player2_name():
    global player2_name
    player2_name = input("Player 2, please enter your name: ")
    return player2_name

def get_player1_selection():
    player1_input = getpass.getpass(f"{player1_name}, please enter a choice (rock[0], paper[1], scissors[2]): ")
    selection = int(player1_input)
    action = Action(selection)
    return action

def get_player2_selection():
    player2_input = input(f"{player2_name}, please enter a choice (rock[0], paper[1], scissors[2]): ")
    selection = int(player2_input)
    action = Action(selection)
    return action

def validcheck_player1(player1_input):
    if player1_input == Action.rock:
        print("Valid move")
    elif player1_input == Action.paper:
        print("Valid move")
    elif player1_input == Action.scissors:
        print("Valid move")
    else:
        raise Exception("Invalid moves")

def validcheck_player2(player2_input):
    if player2_input == Action.rock:
        print("Valid move")
    elif player2_input == Action.paper:
        print("Valid move")
    elif player2_input == Action.scissors:
        print("Valid move")
    else:
        raise Exception("Invalid moves")

def roundwinner():
    
    player1_score = 0
    player2_score = 0
    
    while True:
        
        try:
            player1_input = get_player1_selection()
            validcheck_player1(player1_input)
        except:
            print("Exception: Player 1 invalid move")
            continue
        
        try:
            player2_input = get_player2_selection()
            validcheck_player2(player2_input)
        except:
            print("Exception: Player 2 invalid move")
            continue
            
        if player1_input == player2_input:
            print("This round is a tie")
        elif player1_input == Action.rock:
            if player2_input == Action.scissors:
                print(f"{player1_name} wins the round")
                player1_score += 1
            elif player2_input == Action.paper:
                print(f"{player2_name} wins the round")
                player2_score += 1
        elif player1_input == Action.paper:
            if player2_input == Action.rock:
                print(f"{player1_name} wins the round")
                player1_score += 1
            elif player2_input == Action.scissors:
                print(f"{player2_name} wins the round")
                player2_score += 1
        elif player1_input == Action.scissors:
            if player2_input == Action.paper:
                print(f"{player1_name} wins the round")
                player1_score += 1
            elif player2_input == Action.rock:
                print(f"{player2_name} wins the round")
                player2_score += 1

        print(f"{player1_name}'s score is {player1_score} and {player2_name}'s score is {player2_score}")
        
        if player1_score == 2:
            print(f"{player1_name} has won!")
            break
        elif player2_score == 2:
            print(f"{player2_name} has won!")
            break
        else:
            if player1_score > player2_score:
                print(f"{player1_name} is in the lead")
            elif player1_score < player2_score:
                print(f"{player2_name} is in the lead")
            else:
                print(f"{player1_name} and {player2_name} are tied")

def main(): 
    while True:
        player1_name = get_player1_name()
        player2_name = get_player2_name()
        roundwinner()
        while True:
            restart = input('New game? (y/n): ').lower()
            if restart == 'n':
                print("Game over")
                exit()
            elif restart == 'y':
                break
            else:
                print("Invalid input")
                continue
    
if __name__ == "__main__":
    main()
```
